```python
data1 = pd.read_csv('forbes_2022_billionaires.csv', sep=',')
```

```python
data1 = data1.dropna(axis=0, how='any')
data1.shape
data1
```

```python
data1.info()
```

```python
#Кодирование категориальных признаков
category_cols = ['personName', 'category', 'source', 'country', 'state', 
                 'city', 'countryOfCitizenship', 'organization', 'gender', 
                'birthDate', 'title', 'residenceMsa', 'bio', 'about']

print('Количество уникальных значений\n')
for col in category_cols:
    print(f'{col}: {data1[col].unique().size}')
```

```python
remove_cols = ['personName', 'category', 'source', 'country', 'state', 
                 'city', 'countryOfCitizenship', 'organization', 'gender', 
                'birthDate', 'title', 'residenceMsa', 'bio', 'about']
```

```python
for col in remove_cols:
    category_cols.remove(col)
data1 = pd.get_dummies(data1, columns=category_cols)
```

```python
data1.drop(remove_cols, axis=1, inplace=True)
data1.describe()
```

```python
#Разделение выборки на обучающую и тестовую
y = data1['finalWorth']
X = data1.drop('finalWorth', axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
#Масштабирование данных
scaler = MinMaxScaler().fit(x_train)
x_train = pd.DataFrame(scaler.transform(x_train), columns=x_train.columns)
x_test = pd.DataFrame(scaler.transform(x_test), columns=x_train.columns)
x_train.describe()
```

```python
#Обучение KNN с произвольным k
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
def print_metrics(y_test, y_pred):
    print(f"R^2: {r2_score(y_test, y_pred)}")
    print(f"MSE: {mean_squared_error(y_test, y_pred)}")
    print(f"MAE: {mean_absolute_error(y_test, y_pred)}")
    
def print_cv_result(cv_model, x_test, y_test):
    print(f'Optimizacia metrici {cv_model.scoring}: {cv_model.best_score_}')
    print(f'Lyshiy parametr {cv_model.best_params_}')
    print(f'Metricic na testovom nabore')
    print_metrics(y_test, cv_model.predict(x_test))
    print()
    
```

```python
base_k = 3
base_knn = KNeighborsRegressor(n_neighbors=base_k)
base_knn.fit(x_train, y_train)
y_pred_base=base_knn.predict(x_test)
print(f'Test metrics for knn with k={base_k}\n')
print_metrics(y_test, y_pred_base)
```

```python
#Кросс-валидация
from sklearn.model_selection import GridSearchCV
metrics = ['r2', 'neg_mean_squared_error', 'neg_mean_absolute_error']
cv_values=[2, 3]

for cv in cv_values:
    print(f'Результаты кросс-валидации при cv={cv}\n')
    for metric in metrics:
        params= {'n_neighbors': range(1, 30)}
        knn_cv = GridSearchCV(KNeighborsRegressor(), params, cv=cv, scoring=metric, n_jobs=-1)
        knn_cv.fit(x_train, y_train)
        print_cv_result(knn_cv, x_test, y_test)
```

```python
best_k = 3
y_pred_best = KNeighborsRegressor(n_neighbors=best_k).fit(x_train, y_train).predict(x_test)
```

```python
#Сравнение исходной и оптимальной моделей
print('Исходная модель\n')
print_metrics(y_test, y_pred_base)
print('_______________________')
print('\nОптимальная модель\n')
print_metrics(y_test, y_pred_best)
```
